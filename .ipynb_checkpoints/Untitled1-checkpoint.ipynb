### Count plots

```python
batch1 = ["Geography","Gender","Age","Tenure"] # Deklarasi Data Kategor Batch 1("Geography,Gender,Age,Tenure")
batch2 = ["NumOfProducts","HasCrCard","IsActiveMember","Exited"] # Deklarasi Data Kategori Batch 2("NumOfProducts,HasCrCard,IsActiveMember,Exited")
```

```python
#Membuat CountPlots untuk data Ketegirikal "Batch1"
for column in batch1:
    plt.figure(figsize=(15,3))
    plt.title(f"{column}")
    sns.countplot(bank[column])
    plt.show()
```

```python
#Membuat CountPlots untuk data Ketegirikal "Batch2"
for column in batch2:
    plt.figure(figsize=(15,3))
    plt.title(f"{column}")
    sns.countplot(bank[column])
    plt.show()
```

* Kolom "Age" dimana Umur 37 lah yang paling Terbanyak
* Kolom "Tenure" dimana 2 lah data yang paling terbanyak




Pada Data Kategorikal Batch 2 :

## DATA NUMERIKAL

### Unvariate Analysis

```python
# Setting up subplots
f,axes=plt.subplots(ncols=3, nrows=3, figsize=(15,5), sharex=True)

column=['NumOfProducts','HasCrCard','IsActiveMember']
for i in range(3):
    
    #Ploting the boxplot
    sns.boxplot(x = bank[column[i]], ax=axes[i][0], color='powderblue')
    
    #Plotting With Distribution Plots
    sns.histplot(data = bank[column[i]], ax=axes[i][1],  color="red")# Ploting Histogram        
    sns.histplot(data = bank[column[i]], ax=axes[i][2],  color="red", kde=True, stat="density", linewidth=0)  # Ploting the Histogram with KDE     
    
    plt.tight_layout()
```

## Bivariate Analysis

```python
bank["Exited"].unique()
```

```python
active = bank.loc[bank["Exited"] == 1]
non_active = bank.loc[bank["Exited"] == 0] 
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 1,figsize=(15, 5), sharex=True)
sns.despine(left=True)  #sepal_length-vs-class

# Ploting the histogram with KDE

sns.histplot(data = active["CreditScore"],label="active-Credits",color='b',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data = non_active["CreditScore"],label="nonactive-Credits",kde=True,color='orange',linewidth=2,alpha=0.3)

plt.legend()
plt.tight_layout()
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 1,figsize=(15, 5), sharex=True)
sns.despine(left=True)  #sepal_length-vs-class

# Ploting the histogram with KDE

sns.histplot(data = active["Age"],label="active-Age",color='b',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data = non_active["Age"],label="nonactive-Age",kde=True,color='orange',linewidth=2,alpha=0.3)

plt.legend()
plt.tight_layout()
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 1,figsize=(15, 5), sharex=True)
sns.despine(left=True)  #sepal_length-vs-class

# Ploting the histogram with KDE

sns.histplot(data = active["Balance"],label="active-Balance",color='b',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data = non_active["Balance"],label="nonactive-Balance",kde=True,color='orange',linewidth=2,alpha=0.3)

plt.legend()
plt.tight_layout()
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 1,figsize=(15, 5), sharex=True)
sns.despine(left=True)  #sepal_length-vs-class

# Ploting the histogram with KDE

sns.histplot(data = active["EstimatedSalary"],label="active-EstimatedSalary",color='b',kde=True,linewidth=2,alpha=0.3)
sns.histplot(data = non_active["EstimatedSalary"],label="nonactive-EstimatedSalary",kde=True,color='orange',linewidth=2,alpha=0.3)

plt.legend()
plt.tight_layout()
```
